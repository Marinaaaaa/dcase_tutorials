```python
### Extracting mel-band energies for the audio signal (using default parameters)

# Get audio in a container, mixdown of a stereo signal
audio_container = du.containers.AudioContainer().load(
    filename=du.utils.Example.audio_filename()
).mixdown()

# Create extractor
mel_extractor = du.features.MelExtractor()
mel_extractor.show()

# Extract features
mels = mel_extractor.extract(audio_container)
```

```python
# Extract features directly from numpy matrix:

# Create an audio signal
t = np.linspace(0, 2, 2 * 44100, endpoint=False)
x1 = np.sin(220 * 2 * np.pi * t)

# Create extractor
mel_extractor = du.features.MelExtractor()

# Extract features
mels = mel_extractor.extract(x1)

# Plot
du.containers.DataMatrix2DContainer(
    data=mels,
    time_resolution=mel_extractor.hop_length_seconds
).plot()
```
